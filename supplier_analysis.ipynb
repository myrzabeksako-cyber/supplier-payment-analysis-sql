```python
# Quick start in Python
import pandas as pd
import sqlite3

df_s = pd.read_csv(r"C:\Users\ASUS\Downloads\CONTRACTS.csv")
df_c = pd.read_csv(r"C:\Users\ASUS\Downloads\SUPPLIERS.csv")

conn = sqlite3.connect(':memory:')
df_s.to_sql('SUPPLIERS', conn, index=False)
df_c.to_sql('CONTRACTS', conn, index=False)

# Run any query
result = pd.read_sql("SELECT * FROM SUPPLIERS ORDER BY CONTRACT_SUM DESC", conn)
print(result)
```
